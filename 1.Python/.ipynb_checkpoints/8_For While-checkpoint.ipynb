## For - While

For loop is used in python for sequential traversal or itervative based loops.

The for statement loops are based on a collection of items, putting each item into a temporary variable for use. We'll also need to use the in statement for getting the values in the data structure

While loop will be executed its code block over and over again until its conditions are false. This can be useful if something needs to be repeated over and over for a said amount of time


```python
# range function in Python takes the values from the start index till the end index exclusing the end index
# ex: range(6, 9) gives hte values 6,7,8

for i in range(10):
    print(i)

for i in range(1000):
    i=i+10
    print(i)
```

```python
# third parameter in range function denotes the interval of selecting the values in the index range
# ex: range(0, 7, 2) gives values 0,2,4,6
#ex: range(4, 14, 3) gives values 4,7,10,13

for i in range(1,10,3):
    print(i ** 2) 
```

```python
# for Loop on a string

sample_string = "Is Python a snake?"
b = sample_string.split()
print(b)
for i in b:
    print(i)
```

```python
str1=''
for i in range(0,9):
    if i<5:
        str1 += '* '
        print(str1)
    elif i>4:
        str1 = str1[:-2]
        print(str1)
```

```python
# enumerate iterate through the characters
# n here denotes the index of hte values
for n,alphabet in enumerate(sample_string):
    print(alphabet, n)
```

```python
# find out if a character is vowel

vowels =''
for alphabet in sample_string:
    if alphabet in 'aeiouI':
        vowels += ' ' + alphabet
print(vowels)
```

```python
# Find out the even digits and odd digits in a number

number = '639528418270840236326229'
even_digits=''
odd_digits = ''
for digit in number:
    if int(digit)%2 == 0:
        even_digits += digit
    else:
        odd_digits += digit

print('Even digits are : '+ even_digits + ' & odd digits are : '+ odd_digits)
```

```python
list_of_inventories = ['Apple', 'Banana', 'Potato', 'cherry','Mango', 'Onion', 'Toothpaste']
fruits = ['Apple','Banana','Mango','Orange','Strawberry']
vegetables = ['Potato', 'Onion','Cucumber', 'Celery']
```

```python
#count of fruits in inventory list
count_fruits = 0
#count of vegetables in inventory list
count_veg = 0
for item in list_of_inventories:
    print(item)
    if item in fruits:
        count_fruits+=1
    elif item in vegetables:
        count_veg +=1
    else:
        continue
print(count_fruits)
print(count_veg)
```

```python

#sentence check
latest_tweet = '@imVkohli scored a winning knock to take the ##rcb into play-offs'

for word in latest_tweet.split():
    if word.startswith('@'):
        print('Player Name: '+word[1:])
    elif word.startswith('#'):
        print('Team Name: '+word[2:])
    else:
        print('No player or team is part of the statement')
```



## While loop


```python
x = 0
while x != 10:
    x += 1
    print(x)
```

```python
#while loop
start =30
total= 0
while start<62:
    total+=start    # total = 90+30 = 30
    start+=1        # start = 31+1 =31
    print(start)

print(total)

```

```python
#break ststement

a = 0
while True:
    print(a)
    a += 1
    if a == 10:
        break
```

```python
#continue statement

b = 0
while b != 10:
    b += 1
    if b < 5:
        continue
    print(b)
```
